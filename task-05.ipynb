## Importing necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Importing and Viewing Dataset

```python
df=pd.read_csv('/kaggle/input/us-accidents/US_Accidents_March23.csv')
```

### Pre-processing

```python
df.duplicated().sum()
```

```python
df=df.dropna(subset=['Precipitation(in)'])
```

```python
df=df.dropna(subset=['Temperature(F)','Wind_Chill(F)','Humidity(%)','Pressure(in)','Visibility(mi)','Wind_Direction', 'Wind_Speed(mph)',
                      'Weather_Condition'])
```

```python
df.isna().sum()/len(df)*100
```

```python
df=df.dropna(subset=['City','Sunrise_Sunset',
       'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight'])
```

```python
df.isna().sum()/len(df)*100
```

## Top states W.R.T accident count

```python
states = df.State.value_counts().head()
plt.figure(figsize=(10,7))
sns.barplot(y=states, x=states.index)
plt.title('States with highest number of accidents')
plt.show()
```

## Top cities based on Accident count

```python
cities=df.City.value_counts().head()
plt.figure(figsize=(10,7))
sns.barplot(x=cities.values, y=cities.index)
plt.title('Cities with highest number of accidents')
plt.show()
```

```python
df.Start_Time=pd.to_datetime(df.Start_Time, errors='coerce')
```

## Visualizing trends in accidents over time

### In general

```python
plt.figure(figsize=(10,5))
sns.barplot(x=df.Start_Time.dt.hour.value_counts().index,
            y=df.Start_Time.dt.hour.value_counts().values/len(df)*100)
plt.title('Accidents over time')
plt.show()
```

### Over days

```python
plt.figure(figsize=(10,5))
sns.barplot(x=df.Start_Time.dt.day_of_week.value_counts().index,
            y=df.Start_Time.dt.day_of_week.value_counts().values/len(df)*100)
plt.title("Accidents over days")
plt.show()
```

```python
weekends_data = df[(df['Start_Time'].dt.day_of_week == 5) | (df['Start_Time'].dt.day_of_week==6)]
```

### Specifically on weekends

```python
plt.figure(figsize=(10,5))
sns.barplot(x=weekends_data.Start_Time.dt.hour.value_counts().index,
            y=weekends_data.Start_Time.dt.hour.value_counts().values)
plt.title('Accidents over time on weekends')
plt.show()
```

### Month Based

```python
df.Start_Time.dt.month.value_counts().sort_index().plot(kind='bar')
plt.title('Accidents by month')
plt.show()
```

### Over the years

```python
sns.countplot(x=df.Start_Time.dt.year, data=df)
plt.title('Accidents over the years')
plt.show()
```

### Accidents by time of the day

```python
fig, ax = plt.subplots(figsize=(8,5))
sns.histplot(df['Start_Time'].dt.hour, bins = 24)

plt.xlabel("Start Time")
plt.ylabel("Number of Occurence")
plt.title('Accidents Count By Time of Day')
plt.show()
```

## Severity of Accidents

```python
plt.pie(df.Severity.value_counts(),labels=df.Severity.value_counts().index,autopct='%1.1f%%')
plt.title('Accidents with respect to Severity')
plt.show()
```

```python
df.Source.value_counts().plot(kind='bar')
plt.show()
```

```python
weather_condition = df.Weather_Condition.value_counts()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
weather_condition.sort_values(ascending=False)[:20].plot(kind='bar')
ax.set(title = 'Weather Conditions at Time of Accident Occurence',
       xlabel = 'Weather',
       ylabel = 'Accidents Count')
plt.show()
```
